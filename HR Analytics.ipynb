```python
# b) Distribution of our Labels
sns.countplot(x='Attrition', data=df)
plt.show()
```

```python
# II. Gender Analysis:
# a) Age Distribution by Gender
fig = px.histogram(df, x='Age', color='Gender', nbins=20, title='Age Distribution by Gender')
fig.show()
```

```python
# b) Job Satisfaction Distribution by Gender
fig = px.histogram(df, x='JobSatisfaction', color='Gender', nbins=20, title='Job Satisfaction Distribution by Gender')
fig.show()
```

```python
# c) Monthly Income by Gender
fig = px.box(df, x='Gender', y='MonthlyIncome', title='Monthly Income by Gender')
fig.show()
```

```python
# d) Presence by Department
fig = px.histogram(df, x='Department', color='Gender', title='Presence by Department')
fig.show()
```

```python
# IV. The Impact of Income towards Attrition:
# b) Determining Satisfaction by Income
fig = px.scatter(df, x='MonthlyIncome', y='JobSatisfaction', title='Determining Satisfaction by Income')
fig.show()
```

```python
# c) Income and the Levels of Attrition
fig = px.box(df, x='Attrition', y='MonthlyIncome', title='Income and the Levels of Attrition')
fig.show()
```

```python
# e) Level of Attrition by Overtime
fig = px.histogram(df, x='OverTime', color='Attrition', title='Level of Attrition by Overtime')
fig.show()
```

```python
# V. Working Environment:
# a) Number of Employees in each Job Role
fig = px.histogram(df, y='JobRole', title='Number of Employees in each Job Role')
fig.show()
```

```python

# c) Attrition by Job Role
fig = px.histogram(df, y='JobRole', color='Attrition', title='Attrition by Job Role')
fig.show()
```

```python
# VII. Analysis and Models
# a) Correlation Matrix
corr_matrix = df.corr()
fig = px.imshow(corr_matrix, x=corr_matrix.columns, y=corr_matrix.columns, title='Correlation Matrix')
fig.show()
```
